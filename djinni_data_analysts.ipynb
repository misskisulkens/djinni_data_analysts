# Scraping the Data

!!! this code might need editing later on since Djinni changes their HTLM almost every day !!!

```python
import requests as r
from bs4 import BeautifulSoup as bs
import pandas as p
```

```python
initial_link = "https://djinni.co/developers/?title=Data+Analyst"
response = r.get(initial_link)
soup_response = bs(response.text, "html.parser")
response
```

```python
links = []
link_beginning = "https://djinni.co/developers/?title=Data+Analyst&page="
pages = []
page_raw = soup_response.find_all("a", class_ = "page-link")
for page in page_raw[1:-1]:
    pages.append(page.text.strip())
max_page = max(pages)
number = 0
for item in range(int(max_page)):
    number += 1
    new_link = f"{link_beginning}{str(number)}"
    links.append(new_link) 
```

```python
titles = []
salaries = []
details_lists = []
for link in links:
    response = r.get(link)
    soup_response = bs(response.text, "html.parser")
    candidates_data = soup_response.find_all("div", class_ = "candidate-header d-flex flex-column flex-sm-row mt-3")
    for candidate_data in candidates_data:
        title = candidate_data.find("a", class_ = "profile").text.strip()
        salary = candidate_data.find("h3", class_ = "text-success mb-2 mt-0").text.strip()
        details = candidate_data.find("p", class_ = "text-gray-600 d-flex align-items-center flex-wrap mb-4").find_all("span")
        details_list = []
        for detail in details:
            details_list.append(detail.text.strip())
        titles.append(title)
        salaries.append(salary)
        details_lists.append(details_list)
        
```

```python
countries = []
experience_list = []
english_list = []
statuses = []
for dl in details_lists:
    while "·" in dl:
        dl.remove("·")
    countries.append(dl[0])
    experinece = ""
    english = ""
    if len(dl) == 4:
        experinece = dl[1]
        english = dl[2]
    elif len(dl) == 5 and dl[-1] == "У пасивному пошуку":
        experinece = dl[1]
        english = dl[2]
    else:
        experinece = dl[2]
        english = dl[3]
    experience_list.append(experinece)
    english_list.append(english)
    active = True
    if dl[-1] == "У пасивному пошуку":
        active = False
    else:
        active = True
    statuses.append(active)
    print(dl)
```

```python
experience_list_updated = []
for el in experience_list:
    experience_month = 0
    if el == "Без досвіду":
        experience_month = 0
    elif el == "6 місяців":
        experience_month = 6
    elif el == "1.5 роки досвіду":
        experience_month = 18
    elif el == "2.5 роки досвіду":
        experience_month = 30
    elif el == "10 років досвіду" or el == "Більше 10 років досвіду":
        experience_month = 120
    else:
        experience_month = int(el[0])*12
    experience_list_updated.append(int(experience_month))
```

```python
experience_list_updated[:6]
```

```python
candidates_data = []
index = -1
for item in range(len(titles)):
    index += 1
    candidate_dictionary = {"title": titles[index],
                            "country": countries[index],
                            "salary": int(salaries[index].replace("$", "")),
                            "experience": experience_list_updated[index],
                            "english": english_list[index],
                            "status": statuses[index]}
    candidates_data.append(candidate_dictionary)
```
